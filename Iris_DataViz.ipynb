```python
#DATASET IRIS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
#VISUALIZACAO DOS DADOS COM HISTOGRAMAS
```

```python
sns.set(style="darkgrid")
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
c = sns.FacetGrid(a, col="species")
c.map(plt.hist, "petal_length")
```

```python
plt.scatter(np.array(iris.data[:50,0]), np.array(iris.data[:50,2]), marker='o', label='setosa')
plt.scatter(np.array(iris.data[50:,0]), np.array(iris.data[50:,2]), marker='x', label='versicolor')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend()
plt.show()
```
