```python
#DATASET IRIS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
sns.set(style="darkgrid")
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")
b.map(plt.hist, "sepal_length")
```

```python
c = sns.FacetGrid(a, col="species")
c.map(plt.hist, "petal_length")
```

```python
plt.scatter(np.array(iris.data[:50,0]), np.array(iris.data[:50,2]), marker='o', label='setosa')
plt.scatter(np.array(iris.data[50:,0]), np.array(iris.data[50:,2]), marker='x', label='versicolor')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend()
plt.show()
```

```python
plt.scatter(np.array(iris.data[:50,0]), np.array(iris.data[:50,2]), marker='o', label='setosa')
plt.scatter(np.array(iris.data[100:,0]), np.array(iris.data[100:,2]), marker='x', label='viriginica')
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.legend()
plt.show()
```

```python
#df.rename(columns={0:'sepal_length', 1:'sepal_width', 2:'petal_length', 3:'petal_width', 4:'species'})
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
#sns.pairplot(df[['sepal_length', 'sepal_width', 'petal_length',
#        'petal_width','species']], hue='species')
sns.pairplot(iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
        'petal width (cm)','species']], hue='species')
```
